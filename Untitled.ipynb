```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot

faceCascade= cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
vid= cv2.VideoCapture(0)
while True:
        ret,frame=vid.read()
        answer = DeepFace.analyze(frame, actions=['emotion'],enforce_detection=False)
        sanswer = DeepFace.analyze(frame,actions=['age'],enforce_detection=False)
        
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        faces= faceCascade.detectMultiScale(gray,1.1,4)
        
        
        for(x,y,w,h) in faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(103, 14, 212),2)
            
            font= cv2.FONT_HERSHEY_SIMPLEX
            
            cv2.putText(frame,
                       answer['dominant_emotion'],
                       (x,y-20),font,1,
                       (103, 14, 212),2,cv2.LINE_4)
            cv2.putText(frame,str(sanswer['age']),(x,y-60),font,1,
                       (103, 14, 212),2,cv2.LINE_4)
            cv2.imshow('original video',frame)
            
            if cv2.waitKey(1) & 0xFF==ord('k'):
                break
                
                
vid.release()
cv2.destroyAllWindows()
    
```
